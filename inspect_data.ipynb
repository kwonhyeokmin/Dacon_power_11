```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
from matplotlib import pyplot as plt
import matplotlib
```

```python
ROOT_DIR = os.getcwd()
DATA_DIR = os.path.join(ROOT_DIR, 'data')
```

```python
dataset_train = pd.read_csv(os.path.join(DATA_DIR, 'train.csv'))
dataset_test = pd.read_csv(os.path.join(DATA_DIR, 'test.csv'))
submission = pd.read_csv(os.path.join(DATA_DIR, 'submission_1002.csv'))
```

## Train dataset

```python
dataset_train.head(5)
```

```python
print("{start_time} 부터 {end_time} 까지 전력 데이터".format(start_time=min(dataset_train['Time']), end_time=max(dataset_train['Time'])))
dataset_train.describe()
```

```python
# Time column(str type) datetime 객체로 변경
import datetime

dataset_train['Time'] = pd.Series([datetime.datetime.strptime(x, "%Y-%m-%d %H:%M") for x in dataset_train['Time']])
```

```python
dataset_train.head(5)
```

```python
print("미터기 수: {}개".format(len(dataset_train.columns) - 1))
print('미터기는 {}부터 {}까지 입니다.'.format(min(dataset_train.columns[1:], key=lambda x: int(x.replace('X', ''))), 
                                   max(dataset_train.columns[1:], key=lambda x: int(x.replace('X', '')))))
```

```python
# null 개수
null_count = [dataset_train[col].isnull().sum() for col in dataset_train.columns[1:]]
print('null 개수 최소값: {}'.format(np.min(null_count)))
print('null 개수 최대값: {}'.format(np.max(null_count))) 
print('null 개수 평균값: {}'.format(np.mean(null_count)))
print('총 row 개수: {}'.format(len(dataset_train)))
```
